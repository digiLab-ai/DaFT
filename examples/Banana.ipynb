```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from daft import DaFT
```

```python
np.random.seed(123)
```

```python
def get_banana(X, a=2, b=0.5):
    X[:,0], X[:,1] = a*X[:,0], X[:,1]/a + b*(X[:,0]**2 + a**2)
    return X
```

```python
n_samples = 10000
n_sub = 20

n_dim = 2
n_features = 2000
```

```python
X = get_banana(np.random.normal(loc=0, scale=1, size=(n_samples, n_dim)))
```

```python
Y_random = X[np.random.randint(X.shape[0], size=n_sub), :]
```

```python
my_daft = DaFT(X, n_sub, multiprocessing=False)
Y_best = my_daft.run(n=500, ngen=200)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0], X[:,1], alpha=0.1, c='k', label='Complete')
plt.scatter(Y_random[:,0], Y_random[:,1], alpha=1, c='tomato', label='Random')
plt.scatter(Y_best[:,0], Y_best[:,1], alpha=1, c='springgreen', label='DaFT')
plt.legend()
#plt.savefig('banana.png', bbox_inches='tight')
plt.show()
```
